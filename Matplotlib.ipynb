```python
import numpy as np
x=np.array([1, 2, 3, 4, 5])
y = 2*(x**4)+100
plt.plot(x, y)
plt.show()
```

```python
x=np.arange(0, 10, 0.1)
y = 2*(x**4)+100
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.25) #This accepts three arguments, start index(optional), end index and step count.
y = x**4

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.25) #This accepts three arguments, start index(optional), end index and step count.
y = x**4

plt.plot(x, y, color = "red")
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "p")
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+")
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+", linewidth="3") #linewidth can also be written without double quotes.
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+", label="x^3")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+", label="x^3")
plt.legend()
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+", label="x^3")
plt.legend()
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.title("Matplotlib Chart")
plt.show()
```

```python
x = np.arange(0, 10, 0.25)
y = x**4

plt.plot(x, y, color = "red", marker = "+", label="x^3")
plt.legend()
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.title("Matplotlib Chart")
plt.axis([0, 15, 0, 15000]) #This signifies that x asis is starting from 0 and ending at 14(15 exclusive). And similarly for y axis.
plt.text(10, 12500, "Radhika", fontsize=15)
plt.show()
```

```python
profit = [10, 20, 30, 40, 50]
lables = ["Google", "Amazon", "Facebook", "Flipkart"]

plt.prism()
plt.show()
```

```python
profit = [10, 20, 30, 40, 50]
lables = ["Google", "Amazon", "Facebook", "Flipkart", "Microsoft"]

plt.pie(profit)
plt.show()
```

```python
colors = ["Red", "Magenta", "Blue", "Green", "Purple"]
plt.pie(profit, colors= colors)
```

```python
plt.pie(profit, colors= colors, labels=lables)
```

```python
plt.pie(profit, colors= colors, labels=lables, autopct="%.2f%%")
```

```python
explode=[0.05, 0.1, 0, 0, 0]
plt.pie(profit, colors= colors, labels=lables, autopct="%.2f%%", explode=explode)
```

```python
plt.pie(profit, colors= colors, labels=lables, autopct="%.2f%%", explode=explode, counterclock=False)
```

```python
plt.pie(profit, colors= colors, labels=lables, autopct="%.2f%%", explode=explode, counterclock=False, startangle=90)
```
