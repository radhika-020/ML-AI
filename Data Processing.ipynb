```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("Covid_Data_new - Covid_Data_new.csv")
x = data_set.iloc[:, :-1].values #This will print all the rows but columns will be printed till the second last as the last one is excluded.
print(x) #Note that entries saying 'nan' are entries which were empty initially but pandas added nan in place of empty values.
```

```python
y = data_set.iloc[:, -1].values
print(y)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 0:5])
print(x) #This line gives an error because the columns 1, 2 and 3 are string type, therefore they cannot be implemented as the mean strategy. 
#sklearn library checks all the missing values and replaces them with the mean of the full column.
```

```python
imputer.fit(x[:, 0:1])
print(x) 
```

```python
imputer.fit(x[:, 0:1])
x[:, 0:1] = imputer.transform(x[:, 0:1])
print(x) 
```

```python
imputer.fit(x[:, 4:5])
x[:, 4:5] = imputer.transform(x[:, 4:5])
print(x) 
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 0:5:4])
x[:, 0:5:4] = imputer.transform(x[:, 0:5:4])
print(x)
```
