```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Imporing data set
#Simple Linear Regression is dependent on a single independent variable. The equation for the same can be written as y = ax1 + c, here y si the dependent variable which is dependent on x1, x1 is the independent variable, and a and c are constants.
data_set = pd.read_csv("Salary_Data.csv")
X = data_set.iloc[:, :-1].values 
y = data_set.iloc[:, -1].values
print(X)
```

```python
#Splitting into test and data set
from sklearn.model_selection import train_test_split #This library is used to train and test data according to the given values. sklearn is also used for regression, clusering, classificaion and other useful ML techniques.
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #This regressor is an insance of this funcion.
regressor.fit(X_train, y_train)  
```

```python
#Basically we are as per the given X_test value, we are predicting the value for y.
y_pred = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regressor.predict(X_train), color="red")
```
